```python
# merged_df_3 = merged_df_2.merge()
def remove_suffix(column_name, suffixes):
    for suffix in suffixes:
        if column_name.endswith(suffix):
            return column_name[: -len(suffix)]
    return column_name


# List of suffixes to remove
suffixes_to_remove = ["_cen", "_zip", "_county"]

# Loop through the columns in df1 and df2 to fill NaN values in df2 with values from df1
# index = 0
for column in merged_df.columns:
    if column in zip_df.columns and column != "ZIPCODE" and column != "ZIP":
        # index += 1
        merged_df_2[column + "_cen"].fillna(merged_df_2[column + "_zip"], inplace=True)
        merged_df_2.drop(columns=[column + "_zip"], inplace=True)
        # if(index %500 =)

merged_df_2.columns = [
    remove_suffix(col, suffixes_to_remove) for col in merged_df_2.columns
]
merged_df_2
```
