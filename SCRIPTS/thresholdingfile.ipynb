```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_c = pd.read_csv('final_feature1.csv')
df_c = df_c[['Variable Label','Risk','Score']]
df_code = pd.read_excel('SDOH_2020_Codebook.xlsx', sheet_name = 'All SDOH Variables')
df_code = df_code[['Variable Name','Variable Label','Topic','Domain']]
df_code.head()
```

```python
df_c
```

```python
df33 = pd.merge(df_code, df_c, on='Variable Label', how='inner')
df33
```

```python
df_final = pd.read_csv('2020_2019_2018_merged_data.csv')
```

```python
dict2 ={
    'Label' : [],
    'Name' : [],
    'Risk' : [],
    'Score' : [],
    '1' : [],
    '2' : [],
    '3' : [],
    '4' : [],
}
```

```python
notpos = []
for i in range(len(df33)):
    r = df33['Variable Name'].to_list()[i]
    try:
        if df33['Score'].to_list()[i] == 1:
            _, bins = pd.qcut(df_final[r], [0,0.6,0.85,1], retbins = True)
            dict2['1'].append(float(bins[0]))
            dict2['2'].append(float(bins[1]))
            dict2['3'].append(float(bins[2]))
            dict2['4'].append(float(bins[3]))
        else:
            _, bins = pd.qcut(df_final[r], [0,0.15,0.4,1], retbins = True)
            dict2['4'].append(float(bins[0]))
            dict2['3'].append(float(bins[1]))
            dict2['2'].append(float(bins[2]))
            dict2['1'].append(float(bins[3]))
        dict2['Label'].append(df33['Variable Label'].to_list()[i])
        dict2['Name'].append(r)
        dict2['Risk'].append(df33['Risk'].to_list()[i])
        dict2['Score'].append(df33['Score'].to_list()[i])
    except:
        notpos.append(i)
        print(df33['Variable Label'].to_list()[i])
        print(r)
        print(df33['Risk'].to_list()[i])
        print(df33['Score'].to_list()[i])
        print('\n\n')
```

```python
error1 = []
for e in notpos:
    r = df33['Variable Name'].to_list()[e]
    try:
        if df_final[r].mode().tolist()[0] == 0:
            lst1 = df_final[r].dropna().tolist()
            lst2 = [x for x in lst1 if x != 0]
            lst2 = np.array(lst2)
            print(np.quantile(lst2, 0.7))
            if df33['Score'].to_list()[e] == 1:
                dict2['1'].append(min(lst1))
                dict2['2'].append(float(0))
                dict2['3'].append(float(np.quantile(lst2, 0.7)))
                dict2['4'].append(max(lst1))
            else:
                dict2['1'].append(max(lst1))
                dict2['2'].append(float(np.quantile(lst2, 0.7)))
                dict2['3'].append(float(0))
                dict2['4'].append(min(lst1))
            dict2['Label'].append(df33['Variable Label'].to_list()[e])
            dict2['Name'].append(df33['Variable Name'].to_list()[e])
            dict2['Risk'].append(df33['Risk'].to_list()[e])
            dict2['Score'].append(df33['Score'].to_list()[e])
    except:
        error1.append(r)
```

```python
pd.DataFrame(dict2).to_csv('threehold_variable.csv')
pd.DataFrame(dict2)
```

```python
df_health = pd.read_csv('healthdata_transformed.csv')
df_health
```

```python
df_health = df_health.iloc[:,2:]
```

```python
# np.quantile(lst2, 0.7)
er = []
lsth = df_health.columns
for i in lsth:
    e = df_health[i].dropna().to_numpy()
    er.append(np.quantile(e, 0.7))
```

```python
df_life = pd.read_csv('life_final.csv')
df_life
```

```python
tq = []
for i in df_life.columns:
    if 'e(x) AT' in i:
        print(i)
        tq.append(i)
```

```python
df_life[tq]
```

```python
# np.quantile(lst2, 0.7)
high = []
low = []
for i in tq:
    e = df_life[i].dropna().to_numpy()
    high.append(np.quantile(e, 0.7))
    low.append(np.quantile(e, 0.3))
```

```python
dict3 = {
    'range' : tq,
    'high' : high,
    'low' : low
}
pd.DataFrame(dict3).to_csv('life_threshold.csv')
```
